```python
# separating the features and target
X = df.drop(columns = 'nchronic', axis = 1)
y = df['nchronic']
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Assuming 'df' is your original DataFrame

# 1. Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()  # Find columns with object dtype

# 2. Create a ColumnTransformer to apply different preprocessing to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X.select_dtypes(exclude=['object']).columns.tolist()), # Apply StandardScaler to numerical columns
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_cols), # Apply OneHotEncoder to categorical columns
    ])

# 3. Fit and transform the data
sdata = preprocessor.fit_transform(X)
```

```python
import warnings 
warnings.filterwarnings('ignore') 
```

```python
#spliting the dataset and training and testing the dataset
# 1. Identify categorical features:
categorical_features = X.select_dtypes(include=['object']).columns # Selects columns with 'object' dtype

# 2. Perform one-hot encoding:
X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=True)
#drop_first=True to avoid multicollinearity


# 3. Re-split the data into training and testing sets:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size = 0.2, random_state = 42)

# 4. Now, train your model:
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
#prediction
y_pred = model.predict(X_test)
print(y_pred)
```

```python
#model performance accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ', accuracy)
```
